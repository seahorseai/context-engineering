```python
# Step 2: Import Required Modules
from langchain import PromptTemplate
from googletrans import Translator
```

```python
# Step 3: Create a PromptTemplate
template = """
You are a helpful assistant that translates English to French.

English: {english_text}
French:
"""

prompt_template = PromptTemplate(
    input_variables=["english_text"],  # List of variables in the template
    template=template
)
```

```python
# Step 4: Fill the Template with Variables
input_variables = {
    "english_text": "Hello, how are you?"
}

prompt = prompt_template.format(**input_variables)

# Print the generated prompt
print("Generated Prompt:")
print(prompt)
```

```python
# Step 5: Use Google Translate for Translation (Optional)
translator = Translator()

# Extract the English text from the prompt
english_text = input_variables["english_text"]

# Perform the translation
translation = translator.translate(english_text, src='en', dest='fr')

# Print the translation
print("\nTranslation:")
print(translation.text)
```
