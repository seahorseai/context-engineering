````python
!pip install -q langchain-openai langchain
from google.colab import userdata
from langchain_openai import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain, SequentialChain

# Set your OpenAI key
OPENAI_API_KEY = userdata.get("OpenAI-key")

# Initialize OpenAI model with API key
llm = OpenAI(api_key=OPENAI_API_KEY, temperature=0)

# Step 1: Code Generation Chain
code_prompt = PromptTemplate(
    input_variables=["instruction"],
    template="Convert the following instruction into a valid Python function named 'generated_function':\n\nInstruction: {instruction}\n\nPython function:\n```python\n"
)
code_generation_chain = LLMChain(llm=llm, prompt=code_prompt, output_key="code") # output_key stores the generated code

# Step 2: Code Explanation Chain
explanation_prompt = PromptTemplate(
    input_variables=["code"],
    template="Explain the following Python function in simple terms:\n\n{code}\n\nExplanation:"
)
explanation_chain = LLMChain(llm=llm, prompt=explanation_prompt, output_key="explanation") # output_key stores the explanation

# Step 3: Utility Chain (Combining Code Generation + Explanation)
utility_chain = SequentialChain(
    chains=[code_generation_chain, explanation_chain], # Define the sequence of chains
    input_variables=["instruction"], # Define the input variable
    output_variables=["code", "explanation"], # Define the output variables
    verbose=True # Enable verbose output for debugging
)

# Example Instruction
instruction = "Write a Python function that takes a list of numbers and returns the sum of all even numbers."

# Run the Utility Chain
output = utility_chain.invoke({"instruction": instruction})

# Extracting Results
print("\n **Generated Python Code:**")
print(output["code"])

print("\n **Code Explanation:**")
print(output["explanation"])
````
